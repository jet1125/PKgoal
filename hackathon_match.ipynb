```python
import numpy as np
from hackathon_qaoa import (
    standard_qaoa,
    ibm_simulator_qaoa,
    ibm_jakarta_qaoa,
    ibm_ku_qaoa,
    ibm_kolkata_qaoa,
)
```

```python
# matching,33 
p = 10

weight = np.array([8, 1, 5, 5, 2, 7, 1, 6, 4])

# 0<=w, w~10^0

mu = np.array([-2*p] * 9) - weight

sigma = np.array(
    [
        [0, p, p, p, 0, 0, p, 0, 0],
        [p, 0, p, 0, p, 0, 0, p, 0],
        [p, p, 0, 0, 0, p, 0, 0, p],
        [p, 0, 0, 0, p, p, p, 0, 0],
        [0, p, 0, p, 0, p, 0, p, 0],
        [0, 0, p, p, p, 0, 0, 0, p],
        [p, 0, 0, p, 0, 0, 0, p, p],
        [0, p, 0, 0, p, 0, p, 0, p],
        [0, 0, p, 0, 0, p, p, p, 0],
    ]
)
```

```python
standard_qaoa(mu, sigma)
```

```python
result_simul_match = ibm_simulator_qaoa(mu, sigma)
result = result_simul_match
```

```python
fval_list = [s.fval for s in result.samples]
import matplotlib.pyplot as plt
import mplhep
plt.style.use(mplhep.style.LHCb2)  # CMS or ATLAS/LHCb

plt.plot([], [], ' ', label='Matching QAOA Simulator')

plt.plot(fval_list, 'o-')
plt.xlabel('$\leftarrow$ Optimized')
plt.ylabel('f Value')
plt.legend()
plt.savefig("MatchQAOASimul.pdf")
plt.show()
```

```python
result_kolkata_match = ibm_kolkata_qaoa(mu, sigma)
result = result_kolkata_match
```

```python
fval_list = [s.fval for s in result.samples]
import matplotlib.pyplot as plt
import mplhep
plt.style.use(mplhep.style.LHCb2)  # CMS or ATLAS/LHCb

plt.plot([], [], ' ', label='Matching QAOA IBMQ')

plt.plot(fval_list, 'o-')
plt.xlabel('$\leftarrow$ Optimized')
plt.ylabel('f Value')
plt.legend()
plt.savefig("MatchQAOAIbmq.pdf")
plt.show()
```
