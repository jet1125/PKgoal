```python
import numpy as np
from hackathon_qaoa import (
    standard_qaoa,
    ibm_simulator_qaoa,
    ibm_jakarta_qaoa,
    ibm_ku_qaoa,
    ibm_kolkata_qaoa,
)
```

```python
# clustering, 2heads, 4 elements
p = 10

weight = np.array([1, 5, 7, 4, 10, 2, 9, 6])

# 0<=w, w~10^0

mu = np.array([-p] * 8) - weight

sigma = np.array(
    [
        [0, 0, 0, 0, p, 0, 0, 0],
        [0, 0, 0, 0, 0, p, 0, 0],
        [0, 0, 0, 0, 0, 0, p, 0],
        [0, 0, 0, 0, 0, 0, 0, p],
        [p, 0, 0, 0, 0, 0, 0, 0],
        [0, p, 0, 0, 0, 0, 0, 0],
        [0, 0, p, 0, 0, 0, 0, 0],
        [0, 0, 0, p, 0, 0, 0, 0],
    ]
)
```

```python
standard_qaoa(mu, sigma)
```

```python
result_simul_cluster = ibm_simulator_qaoa(mu, sigma)
result = result_simul_cluster
```

```python
fval_list = [s.fval for s in result.samples]
import matplotlib.pyplot as plt
import mplhep
plt.style.use(mplhep.style.LHCb2)  # CMS or ATLAS/LHCb

plt.plot([], [], ' ', label='Clustering QAOA Simulator')

plt.plot(fval_list, 'o-')
plt.xlabel('$\leftarrow$ Optimized')
plt.ylabel('f Value')
plt.legend()
plt.savefig("ClusterQAOASimul.pdf")
plt.show()
```

```python
result_kolkata_cluster = ibm_kolkata_qaoa(mu, sigma)
result = result_kolkata_cluster
```

```python
fval_list = [s.fval for s in result.samples]
import matplotlib.pyplot as plt
import mplhep
plt.style.use(mplhep.style.LHCb2)  # CMS or ATLAS/LHCb

plt.plot([], [], ' ', label='Cluster QAOA IBMQ')

plt.plot(fval_list, 'o-')
plt.xlabel('$\leftarrow$ Optimized')
plt.ylabel('f Value')
plt.legend()
plt.savefig("ClusterQAOAIbmq.pdf")
plt.show()
```
